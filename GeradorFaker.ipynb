`tratamentos`

```python
def determinar_probabilidade_gravidade_por_idade_e_periodo(idade, mes_inicio):
    """
    Determina a probabilidade de cada grau de gravidade baseado na idade do paciente
    e no período (mês) de início do tratamento
    """
    # Probabilidades base por idade
    if idade < 5:  # Bebês e crianças pequenas (mais vulneráveis)
        prob_base = [0.25, 0.35, 0.25, 0.15]  # [Leve, Moderada, Grave, Muito Grave]
    elif idade < 18:  # Crianças e adolescentes
        prob_base = [0.50, 0.30, 0.15, 0.05]
    elif idade < 40:  # Adultos jovens
        prob_base = [0.55, 0.30, 0.12, 0.03]
    elif idade < 60:  # Adultos
        prob_base = [0.40, 0.35, 0.20, 0.05]
    else:  # Idosos (mais vulneráveis)
        prob_base = [0.25, 0.35, 0.25, 0.15]

    # Ajuste sazonal - alguns meses têm maior incidência de casos graves
    # Meses de chuva/inverno (maio-setembro) tendem a ter casos mais graves
    if mes_inicio in [5, 6, 7, 8, 9]:  # Período chuvoso/frio
        # Aumenta probabilidade de casos graves
        fator_sazona = [0.9, 1.0, 1.2, 1.3]  # Reduz leve, aumenta grave
    elif mes_inicio in [1, 2, 3, 4]:  # Período de epidemias (verão/outono)
        # Mais casos, mas distribuição mais uniforme
        fator_sazona = [1.1, 1.0, 0.9, 0.8]
    else:  # Período seco
        # Menos casos graves
        fator_sazona = [1.2, 1.0, 0.8, 0.7]

    # Aplica ajuste sazonal
    prob_ajustada = [prob_base[i] * fator_sazona[i] for i in range(4)]

    # Normaliza para somar 1
    total = sum(prob_ajustada)
    prob_final = [p/total for p in prob_ajustada]

    return prob_final
```

```python
def calcular_custo_por_gravidade_e_duracao(gravidade, duracao_dias, idade):
    """
    Calcula o custo do tratamento baseado na gravidade, duração e idade do paciente
    """
    # Custo base diário por gravidade (em reais)
    custo_base_diario = {
        "Leve": 35.0,
        "Moderada": 85.0,
        "Grave": 220.0,
        "Muito Grave": 450.0
    }

    # Fator de ajuste por idade (crianças e idosos custam mais)
    if idade < 5 or idade > 70:
        fator_idade = 1.3
    elif idade < 18 or idade > 60:
        fator_idade = 1.15
    else:
        fator_idade = 1.0

    # Custo diário ajustado
    custo_diario = custo_base_diario[gravidade] * fator_idade

    # Adiciona variação aleatória de ±20%
    variacao = random.uniform(0.8, 1.2)
    custo_diario *= variacao

    # Custo total
    custo_total = custo_diario * duracao_dias

    return round(custo_total, 2)
```

```python
def calcular_duracao_tratamento(gravidade, idade):
    """
    Calcula a duração do tratamento baseado na gravidade e idade
    """
    # Duração base por gravidade (em dias)
    duracao_base = {
        "Leve": (3, 10),
        "Moderada": (7, 21),
        "Grave": (14, 45),
        "Muito Grave": (30, 90)
    }

    min_dias, max_dias = duracao_base[gravidade]

    # Ajuste por idade - crianças pequenas e idosos podem precisar de mais tempo
    if idade < 5 or idade > 70:
        min_dias = int(min_dias * np.random.uniform(0.8, 1.2))
        max_dias = int(max_dias * 1.3)
    elif idade < 18 or idade > 60:
        min_dias = int(min_dias * 1.1)
        max_dias = int(max_dias * 1.25)

    return random.randint(min_dias, max_dias)
```

```python
from datetime import timedelta
```

```python
# Gerar dataset de tratamentos - continuação do código original
tratamentos = []

# Assumindo que os DataFrames pacientes e enfermidades já existem
# (usando os dados do código original)

gravidades_disponiveis = ["Leve", "Moderada", "Grave", "Muito Grave"]
esfermidade_unique = enfermidades['nome'].unique()
# Loop pelos pacientes (continuação do código original)
for _, paciente in pacientes.iterrows():

    # Escolhe enfermidade aleatoriamente
    enfermidade_escolhida = np.random.choice(esfermidade_unique)

    # Pega o enfermidade_id correspondente
    enfermidade_id = enfermidades[enfermidades['nome'] == enfermidade_escolhida]['enfermidade_id'].iloc[0]

    # Calcula idade do paciente
    hoje = pd.to_datetime('today').date()
    idade_paciente = hoje.year - paciente['data_nascimento'].year

    # Gera data de início do tratamento (últimos 2 anos)
    data_inicio = fake.date_between(start_date='-2y', end_date='today')
    mes_inicio = data_inicio.month

    # Determina probabilidade de gravidade baseada na idade e período
    probabilidades = determinar_probabilidade_gravidade_por_idade_e_periodo(idade_paciente, mes_inicio)

    # Escolhe gravidade baseada nas probabilidades
    gravidade_escolhida = np.random.choice(gravidades_disponiveis, p=probabilidades)

    # Calcula duração do tratamento
    duracao_dias = calcular_duracao_tratamento(gravidade_escolhida, idade_paciente)

    # Calcula data fim
    data_fim = data_inicio + timedelta(days=duracao_dias)

    # Calcula custo total
    custo_total = calcular_custo_por_gravidade_e_duracao(gravidade_escolhida, duracao_dias, idade_paciente)

    # Gera ID único para tratamento
    tratamento_id = fake.unique.random_number(digits=10)

    # Pega cidade_id do paciente (assumindo que existe uma correspondência)
    cidade_id = random.randint(1, len(cidades_ceara))

    # Adiciona registro de tratamento
    tratamento = {
        'tratamento_id': tratamento_id,
        'paciente_id': paciente['paciente_id'],
        'enfermidade_id': enfermidade_id,
        'data_inicio': data_inicio,
        'data_fim': data_fim,
        'custo_total': custo_total,
        'cidade_id': cidade_id
    }

    tratamentos.append(tratamento)

# Criar DataFrame de tratamentos
tratamentos_df = pd.DataFrame(tratamentos)
```
